```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
#load dataset
pd.set_option("display.max_columns",None)
df = pd.read_csv("BMW Global Automotive Sales.csv")
df.head(10)
```

```python
#check data type
df.dtypes
```

```python
#fix data types
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df['Month'] = pd.to_numeric(df['Month'], errors='coerce')
df['Units_Sold'] = pd.to_numeric(df['Units_Sold'], errors='coerce')
df['Avg_Price_EUR'] = pd.to_numeric(df['Avg_Price_EUR'], errors='coerce')
df['Revenue_EUR'] = pd.to_numeric(df['Revenue_EUR'], errors='coerce')
df['BEV_Share'] = pd.to_numeric(df['BEV_Share'], errors='coerce')
df['Premium_Share'] = pd.to_numeric(df['Premium_Share'], errors='coerce')
df['GDP_Growth'] = pd.to_numeric(df['GDP_Growth'], errors='coerce')
df['Fuel_Price_Index'] = pd.to_numeric(df['Fuel_Price_Index'], errors='coerce')

df.dtypes
```

```python
#check missing values
df.isna().sum()
```

```python
df_missing = df.copy()
df_missing.loc[0:5, 'Revenue_EUR'] = np.nan   # nan means missing value
df_missing.isna().sum()
```

```python
#remove records
df_removed = df_missing.dropna()
df_removed.shape
```

```python
df_missing.shape
```

```python
#mean
df_imputed_mean = df_missing.copy()
df_imputed_mean['Revenue_EUR'].fillna(df_imputed_mean['Revenue_EUR'].mean(), inplace=True)
df_imputed_mean.head(10)
```

```python

df_imputed_mean.shape
```

```python
df_missing.shape
```

```python
#median
df_imputed_median = df_missing.copy()
df_imputed_median['Revenue_EUR'].fillna(df_imputed_median['Revenue_EUR'].median(), inplace=True)
df_imputed_median.head(10)
```

```python
#outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Revenue_EUR'])
plt.title('Boxplot of Revenue')
plt.show()
```

```python
Q1 = df['Revenue_EUR'].quantile(0.25)
Q3 = df['Revenue_EUR'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['Revenue_EUR'] < lower) | (df['Revenue_EUR'] > upper)]
outliers.head()
```

```python
#remove outliers
df_no_outliers = df[(df['Revenue_EUR'] >= lower) & (df['Revenue_EUR'] <= upper)]
print("Original shape:", df.shape)
print("After removing outliers:", df_no_outliers.shape)
```

```python
#capping
lower_cap = df['Revenue_EUR'].quantile(0.05)
upper_cap = df['Revenue_EUR'].quantile(0.95)
df_capped = df.copy()
df_capped['Revenue_EUR'] = df_capped['Revenue_EUR'].clip(lower_cap, upper_cap)
```

```python
df_capped.shape
```

```python
df[['Revenue_EUR', 'Units_Sold']].head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = df[['Revenue_EUR', 'Units_Sold']].copy()
df_scaled[['Revenue_EUR', 'Units_Sold']] = scaler.fit_transform(df_scaled)
df_scaled.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_standardized = df[['Revenue_EUR', 'Units_Sold']].copy()
df_standardized[['Revenue_EUR', 'Units_Sold']] = scaler.fit_transform(df_standardized)
df_standardized.head()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df_standardized[['Revenue_EUR', 'Units_Sold']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Before PCA)")
plt.show()
```

```python
from sklearn.decomposition import PCA

X = df_standardized[['Revenue_EUR', 'Units_Sold']]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```
